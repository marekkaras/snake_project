```python
import glob
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
```

```python
rewards = glob.glob('./*.csv')
```

```python
data = pd.concat([pd.read_csv(x).rename(columns={'Unnamed: 0': 'iteration'}) for x in rewards])
```

```python
grouped = data.groupby('game_hash')
```

```python
plt.clf()
plt.title('Positive / negative rewards impacting mean score')
plt.xlabel('Number of Games')
plt.ylabel('Mean Score')
legend_labels = []
for hash, g in grouped:
    plt.plot(g['mean_scores'])
    legend_labels.append((g['negative_reward'].values[0], g['positive_reward'].values[0]))
    #print(g)
plt.legend(legend_labels)
plt.show()
```
