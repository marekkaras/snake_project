```python
import glob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
```

```python
rewards = glob.glob('./*.csv')
```

```python
data = pd.concat([pd.read_csv(x).rename(columns={'Unnamed: 0': 'iteration'}) for x in rewards])
```

```python
grouped = data.groupby('game_hash')
```

```python
plt.clf()
plt.title('Game training split')
plt.xlabel('Number of Games')
plt.ylabel('Max Score')
legend_labels = []
for hash, g in grouped:
    #plt.plot(g['mean_scores'])
    #legend_labels.append(('means', g['how_many_seeds'].values[0], g['how_many_games'].values[0], 1000))
    plt.plot(g['scores'])
    legend_labels.append(('maxes', g['how_many_seeds'].values[0], g['how_many_games'].values[0], 1000))
    #print(g)
plt.legend(legend_labels)
plt.show()
```
